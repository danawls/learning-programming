```python
import pandas as pd

df = pd.read_csv("score.csv")
df.set_index("number", inplace=True)
df

def add_num(a):
    return str(a) + "cm"
```

```python
df["Height"] = df["Height"].apply(add_num)
df
```

```python
def get_title(a):
    if pd.notnull(a):
        return a.capitalize()
    return a
df["SW Skill"] = df["SW Skill"].apply(get_title)
df
```

```python
df["School"] = df["School"].str.lower()
df
```
